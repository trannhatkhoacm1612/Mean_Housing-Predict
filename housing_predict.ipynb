# Tiền xử lý (Preprocessing)

Xử lý missing value (Nếu có), có 3 cách để xử lý missing value:
1. Xóa bỏ mẫu đó
2. Xóa bỏ cả đặt trưng đó
3. Điền vào đó là các giá trị như trung vị ,... (thường dùng)

```python
# Kiểm tra xem có missing value không
df.isnull().values.any() # trả về true là có,ngược lại là không
```

```python
# Xem các mẫu có missing value
df[pd.isnull(df).any(axis = 1)]
```

```python
# Thay nhưng giá tị đó bằng trung vị
df.fillna(df.mean(), inplace = True)

df.isnull().values.any()
```

Đưa các dữ liệu dạng hạng mục sang label số, có hai cách:
1. Có thể thay labels theo dạng các số phân biệt (ordinalencoder)
2. Thay giá trị nào có là 1, còn lại 0 (onehot encoder)

```python
convert = pd.get_dummies(df.ocean_proximity,prefix = 'ocean_proximitty')

convert
```

Xóa cột cũ và đưa vào trong bộ dữ liệu các cột vừa xử lý

```python
df.drop(['ocean_proximity'],axis = 1,inplace = True) # xóa cột cũ

df = df.join(convert) # thêm vào

df
```

```python
# Chia tập train và tập test, có thể udnfg hàm split của sklearn
def split_train_test(data,percent):
    shuffle_index = np.random.permutation(len(data)) # hoán vị index của bộ data
    couting_train = int(percent * len(data))
    test_set = shuffle_index[:couting_train]
    train_set = shuffle_index[couting_train:]
    return data.iloc[train_set],data.iloc[test_set]
```

```python
train_set, test_set = split_train_test(df,0.05)
```

```python
print(train_set)
print(test_set)
```

```python
x_train = train_set.drop(['median_house_value'],axis = 1)
y_train = train_set['median_house_value'].copy()
x_test = test_set.drop(['median_house_value'],axis = 1)
y_test = test_set['median_house_value'].copy()
```

# Train

```python
from sklearn.linear_model import LinearRegression

lin = LinearRegression()
lin.fit(x_train,y_train)
```

# Dự đoán

```python
y_pre = lin.predict(x_test)

print('Predict: ', y_pre[:5])
print('Label: ',list(y_test[:5]))
```

# Đánh giá sai số euclid

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test, y_pre))
```
